```python
import transformers
import torch

model_id = "meta-llama/Llama-3.2-1B"
```

```python
# Set up the text generation pipeline
pipeline = transformers.pipeline(
    "text-generation", model=model_id, model_kwargs={"torch_dtype": torch.bfloat16}, device_map="auto"
)

# Define the prompt
prompt = "Once upon a time.."

# Generate the output with increased max_length
generated_output = pipeline(
    prompt,
    max_length=50,  # Increase the length here (this is just an example, you can adjust it further)
    num_return_sequences=1,  # Number of outputs to generate
    no_repeat_ngram_size=2,  # Avoid repeating n-grams (optional)
)

# Print the prompt and output
print(f"Prompt: {prompt}")
print(f"Generated Output: {generated_output[0]['generated_text']}")
```
